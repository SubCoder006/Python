Logging - provide a way to track events, errors and operational information

```python
import logging as log
log.basicConfig(
    filename='newlogs.log',
    filemode='w',
    level=log.DEBUG,
#    level=log.INFO,
#    level=log.WARNING,
#    level=log.ERROR,
#    level=log.CRITICAL,
)
log.debug("This i s a Debug message.")
log.info("This i s a Info message.")
log.warning("This i s a Warning message.")
log.error("This i s a Error message.")
log.critical("This i s a Critical message.")
```
