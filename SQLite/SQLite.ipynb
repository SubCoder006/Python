SQL and SQLite

```python
import sqlite3
```

```python
## connect to sqlite database
connection = sqlite3.connect('example.com')
## create a cursor
cursor = connection.cursor()
```

```python
## creating a table
cursor.execute('''
create table if not exists employees(
    id integer primary key,
    name text not null,
    age integer,
    department text
)               
''')
```

```python
## inserting data
cursor.execute('''
insert into employees(name,age,department)
values('Subayan',20,'AI Developer')
''')

cursor.execute('''
insert into employees(name,age,department)
values('Ravi',24,'ML Developer')
''')

connection.commit()
```

```python
## query the data
cursor.execute('select * from employees')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute('''
update employees
set name = 'Tanishk', age = 23
where id = 2
''')
cursor.execute('''
update employees
set name = 'Baidik', age = 27, department='SDE'
where id = 4
''')
cursor.execute('''
update employees
set name = 'Abdullah', age = 25
where id = 5
''')
```

```python
cursor.execute('select * from employees')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## delete the data
cursor.execute('''
delete from employees
where name = 'Tanishk'               
''')
```

```python
cursor.execute('select * from employees')
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
## close the connection
connection.close()
```
